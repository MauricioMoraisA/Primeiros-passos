```python
#Verificar quantidade de dados faltosos
print(colored('train:\n','red'),train.isnull().sum(),'\n_______-----______\n',colored('test:\n','red'),test.isnull().sum())
```

```python
#vemos que se removermos os dados para treino apenas usaremos 39% dos dados. Porem pela cÃ©lula 
#anterior vemos que a variÃ¡vel objetivo nÃ£o possue valores faltosos, ou seja, podemos fazer um 
# preenchimento com valores existentes do prÃ³prio dataframe 
train.dropna(axis=0).shape[0]/train.shape[0]*100,test.dropna(axis=0).shape[0]/test.shape[0]*100
```

```python
# se assim fizermos nÃ£o zeraremos os valores faltosos. Ainda temos em mente que esse preenchimento nada garante que esses representaram
# os valores reais. No entanto, essa pertubaÃ§Ã£o nos dados tambÃ©m pode contribuir para uma melhor generalizaÃ§Ã£o. assim 
#adotaremos essa tÃ¡tica Juntamente com remoÃ§Ã£o dos valores que ainda permanecerem nan, pois esse sÃ³ preenche com valores anteriores 
# ou posteriores validos e nÃ£o mudaremos o limite de ditanciamento dessa substituiÃ§Ã£o.

train.fillna(method='ffill',inplace=True)
test.fillna(method='ffill',inplace=True)
```

```python
#quantidade de valores faltosos apÃ³s preenchimento
train.isnull().sum(),test.isnull().sum()
```

```python
#removemdo os valores faltosos
train.dropna(inplace=True)
test.dropna(inplace=True)
# verificar quantidade de valores faltosos
train.isnull().sum().sum(),test.isnull().sum().sum()
```
