```python
# defindo pasta raiz
path = 'Desktop/git/input/test.csv'
train = pd.read_csv(r"C:/Users/loveb/Desktop/git/input/train.csv")
test = pd.read_csv(r"C:/Users/loveb/Desktop/git/input/test.csv")
```

```python
# realocando index na base de dados
train.set_index('Date', inplace=True)
test.set_index('Date', inplace=True)
#remover id, não nos servirá de nada
train.drop(['Id'],axis=1,inplace=True)
test.drop(['Id'],axis=1,inplace=True)
```

```python
test.head(2)
```

```python
#transformando variáveis Dummys ( Hot encoding)
train= pd.get_dummies(train, columns=['Location','WindDir9am','WindDir3pm','WindGustDir','RainToday', 'RainTomorrow'], drop_first=True)
test= pd.get_dummies(test, columns=['Location','WindDir9am','WindDir3pm','WindGustDir','RainToday', 'RainTomorrow'], drop_first=True)
```

```python
train.head()
```

```python
#Verificar quantidade de dados faltosos
print(colored('train:\n','red'),train.isnull().sum(),'\n_______-----______\n',colored('test:\n','red'),test.isnull().sum())
```

```python
#vemos que se removermos os dados para treino apenas usaremos 39% dos dados. Porem pela célula 
#anterior vemos que a variável objetivo não possue valores faltosos, ou seja, podemos fazer um 
# preenchimento com valores existentes do próprio dataframe 
train.dropna(axis=0).shape[0]/train.shape[0]*100,test.dropna(axis=0).shape[0]/test.shape[0]*100
```

```python
# se assim fizermos não zeraremos os valores faltosos. Ainda temos em mente que esse preenchimento nada garante que esses representaram
# os valores reais. No entanto, essa pertubação nos dados também pode contribuir para uma melhor generalização. assim 
#adotaremos essa tática Juntamente com remoção dos valores que ainda permanecerem nan, pois esse só preenche com valores anteriores 
# ou posteriores validos e não mudaremos o limite de ditanciamento dessa substituição.

train.fillna(method='ffill',inplace=True)
test.fillna(method='ffill',inplace=True)
```

```python
#quantidade de valores faltosos após preenchimento
train.isnull().sum(),test.isnull().sum()
```

```python
#removemdo os valores faltosos
train.dropna(inplace=True)
test.dropna(inplace=True)
# verificar quantidade de valores faltosos
train.isnull().sum().sum(),test.isnull().sum().sum()
```
