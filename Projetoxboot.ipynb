```python
#vemos que se removermos os dados para treino apenas usaremos 39% dos dados. Porem pela célula 
#anterior vemos que a variável objetivo não possue valores faltosos, ou seja, podemos fazer um 
# preenchimento com valores existentes do próprio dataframe 
train.dropna(axis=0).shape[0]/train.shape[0]*100,test.dropna(axis=0).shape[0]/test.shape[0]*100
```

```python
# se assim fizermos não zeraremos os valores faltosos. Ainda temos em mente que esse preenchimento nada garante que esses representaram
# os valores reais. No entanto, essa pertubação nos dados também pode contribuir para uma melhor generalização. assim 
#adotaremos essa tática Juntamente com remoção dos valores que ainda permanecerem nan, pois esse só preenche com valores anteriores 
# ou posteriores validos e não mudaremos o limite de ditanciamento dessa substituição.

train.fillna(method='ffill',inplace=True)
test.fillna(method='ffill',inplace=True)
```

```python
#quantidade de valores faltosos após preenchimento
train.isnull().sum(),test.isnull().sum()
```

```python
#removemdo os valores faltosos
train.dropna(inplace=True)
test.dropna(inplace=True)
# verificar quantidade de valores faltosos
train.isnull().sum().sum(),test.isnull().sum().sum()
```

```python
cat_vars = train.select_dtypes(include=['object']).columns.tolist()
cat_vars1 = test.select_dtypes(include=['object']).columns.tolist()
```

Agora temos que transformar essas variavéis literias em númericas

```python
#transformando variáveis Dummys ( Hot encoding)
train= pd.get_dummies(train, columns=cat_vars, drop_first=True)
test= pd.get_dummies(test, columns= cat_vars1, drop_first=True)
```

```python
train.iloc[500:600,:].head()
```

```python
#separar treino teste 
xtrain, ytrain= train.iloc[:,:-1].to_numpy(), train.iloc[:,-1].to_numpy()
xtest , ytest =test.iloc[:,:-1].to_numpy(), test.iloc[:,-1].to_numpy()
# normalizar dados
# uma vez que temos uma série temporal multvariada, a normalização mais indicada é a zscore, pois
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
#preserva as caractéristicas originas da série
scaler = StandardScaler()
scaler.fit(xtrain)
xtrain =scaler.transform(xtrain)
scaler.fit(xtest)
xtest = scaler.transform(xtest)
```

```python

import xgboost as xgb

model = xgb.XGBClassifier()

#treinar modelo
model.fit(xtrain, ytrain)
# gerar predicao
ypred = model.predict(xtest)
```

```python
from sklearn.metrics import (recall_score, accuracy_score, precision_score, f1_score, roc_auc_score)

def show_metrics(model, X_test, y_test,y_pred):
    metrics = {
        'recall': recall_score(y_test,y_pred),
        'accuracy': accuracy_score(y_test,y_pred),
        'precision': precision_score(y_test,y_pred),
        'F1': f1_score(y_test,y_pred),
        'roc_auc': roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])
    }
    print(pd.DataFrame.from_dict(metrics, orient='index',columns=['Metrics']))
```

```python
show_metrics(model, xtest, ytest, ypred)
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score, cross_validate, cross_val_predict

folds = 10

kfold = StratifiedKFold(n_splits=folds, shuffle = True, random_state=np.random.randint(200))

model = xgb.XGBClassifier(objective="binary:logistic")
```

```python

# cv_result = cross_val_score(model, X, y, cv=kfold, scoring='roc_auc', verbose=0)
cv_result = cross_validate(model, xtrain, ytrain, cv=kfold, scoring=['recall','accuracy', 'precision', 'f1_macro', 'roc_auc'], verbose=0)
cv_result
```

```python
def show_cv_metrics(str, cv_result):
    print('{}: {:.4f} +/- {:.4f}'.format(str, cv_result[str].mean(), cv_result[str].std()))

show_cv_metrics('test_recall', cv_result)
show_cv_metrics('test_accuracy', cv_result)
show_cv_metrics('test_precision', cv_result)
show_cv_metrics('test_f1_macro', cv_result)
show_cv_metrics('test_roc_auc', cv_result)
```
